# Literature Review

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("data/reg_brak_pm_reduction.xlsx")
data
```

```python
data.sort_values(by=['EV Type', 'PM10 EF Reduction'])
```

```python
# Create a boxplot
fig, ax = plt.subplots(figsize=(5,2))
sns.boxplot(x='PM10 EF Reduction', y='EV Type', data=data, ax=ax, width=0.5, orient='h', order=['HEV', 'PHEV', 'BEV'], palette='Set2')

# Add labels
plt.ylabel('EV Type')
plt.xlabel('PM10 EF Reduction')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(5, 2))
sns.stripplot(x='PM10 EF Reduction', y='EV Type', data=data, order=['HEV', 'PHEV', 'BEV'], orient='h', palette='Set2', size=8, jitter=True, legend=False, dodge=True, alpha=.8)
plt.xlabel('PM10 EF Reduction')
plt.ylabel('EV Type')
plt.show()
```
